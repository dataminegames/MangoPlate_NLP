# 2. Text Preprocessing

### POS tagging
Twitter 형태소 분석기로 형태소 분석

```python
from ckonlpy.tag import Twitter
twitter = Twitter()

def postgging(data):
    tagged_data = []
    for doc in data.values:
        tagged_data.append(twitter.pos(doc))
    return tagged_data
```

```python
tagged_data = postgging(X_train)
```

- 형태소 분석 Test

```python
tagged_data[0][0]
```

### 합성명사 추출

document에서 연속으로 출현하는 명사를 조합해, 합성명사를 사전에 추가

```python
def noun_compound(data):
    josa = list(twitter.dictionary._pos2words['Josa'])

    bigram_compound = set()
    trigram_compound = set()

    for pos, raw in zip(tagged_data, data):
        for i in range(len(pos)-2):
            if pos[i][1] == 'Noun' and pos[i+1][1] == 'Noun':
                string = pos[i][0] + pos[i+1][0]
                if len(re.findall(string, raw)) >= 3 and (pos[i+1][0] not in josa): 
                    bigram_compound.add(string)

    for pos, raw in zip(tagged_data, data):
        for i in range(len(pos)-3):
            if pos[i][1] == 'Noun' and pos[i+1][1] == 'Noun' and pos[i+2][1] == 'Noun':
                string = pos[i][0] + pos[i+1][0] + pos[i+2][0]
                if len(re.findall(string, raw)) >= 3 and (pos[i+2][0] not in josa):
                    trigram_compound.add(string)

    return bigram_compound | trigram_compound
```

```python
compound = noun_compound(X_train)
```

- 합성명사 Test
> 합성 전 / 합성 후 비교

## Cusomized Tokenizer

사전에 등록되지 않은 단어를 추가적으로 형태소 분석

- corpus 생성

```python
#corpus 
corpus = [cor for cor in X_train.values]
corpus[0]
```

- stopwords 리스트 생성

```python
from ckonlpy.utils import load_wordset
stop_words = list(load_wordset('stopwords-ko.txt'))
```

- corpus에서 어미와 조사, stopwords를 제거한 명사 추출

```python
def mytokenizer(raw_sentence):
    tagged_sentences = twitter.pos(raw_sentence)
    tokenized_sentence = []
    for word in tagged_sentences:
        if (word[1] not in ["Eomi", "Josa", "PreEomi"]) and (word[0] not in stop_words):
            tokenized_sentence.append(word[0])           
    return tokenized_sentence
```

- 어미와 조사를 제거한 형태소 분석 test

```python
mytokenizer(corpus[0])
```

# 3. Build Model

### Input Vectors - TF-IDF | Count Vectorizer

- n-gram을 적용한 count 방식 input vector 생성

>1. Unigram Count Vectorizer
2. Bigram Count Vectorizer
3. Trigram Count Vectorizer
4. Unigram TF-IDF Vectorizer
5. Bigram TF-IDF Vectorizer
6. Trigram TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# countvectorizer
c_bow = CountVectorizer(tokenizer=mytokenizer, min_df=30, max_df=0.7, lowercase=False)
c_bow_bigram = CountVectorizer(tokenizer=mytokenizer, min_df=30, max_df=0.7, ngram_range=(1, 2), lowercase=False)
c_bow_trigram = CountVectorizer(tokenizer=mytokenizer, min_df=30, max_df=0.7, ngram_range=(1, 3), lowercase=False)

# tfidfvectorizer
t_bow = TfidfVectorizer(tokenizer=mytokenizer, min_df=30, max_df=0.7, lowercase=False)
t_bow_bigram = TfidfVectorizer(tokenizer=mytokenizer, min_df=30, max_df=0.7, ngram_range=(1, 2), lowercase=False)
t_bow_trigram = TfidfVectorizer(tokenizer=mytokenizer, min_df=30, max_df=0.7, ngram_range=(1, 3), lowercase=False)

# fitting
c_bow_feature = c_bow.fit_transform(corpus)
c_bow_bigram_feature = c_bow_bigram.fit_transform(corpus)
c_bow_trigram_feature = c_bow_trigram.fit_transform(corpus)
t_bow_feature = t_bow.fit_transform(corpus)
t_bow_bigram_feature = t_bow_bigram.fit_transform(corpus)
t_bow_trigram_feature = t_bow_trigram.fit_transform(corpus)
```

```python
# apply to test dataset likewise
tagged_data_test = postgging(X_test)
compound_test = noun_compound(X_test)
try:
    twitter.add_dictionary(compound_test, 'Noun')
except:
    pass
corpus_test = [cor for cor in X_test.values]
c_bow_test = c_bow.transform(corpus_test)
```

```python
cv = KFold(5, shuffle=True)

features = [c_bow_feature, c_bow_bigram_feature, c_bow_trigram_feature, 
            t_bow_feature, t_bow_bigram_feature, t_bow_trigram_feature]

variations = ["Uni_countvectorizer", "Bi_countvectorizer", "Tri_countvectorizer",
             "Uni_TfIdfvectorizer", "Bi_TfIdfvectorizer", "Tri_TfIdfvectorizer"]

target_names = ['bad', 'good', 'missing']
```

6가지 input vector 중 `Unigram Count Vectorizer`를 사용한 모델이 가장 성능이 높게 나왔다.

```python
from sklearn.naive_bayes import MultinomialNB

for i, feature in enumerate(features):
    print("{} starts".format(variations[i]))
    nb_clf= MultinomialNB().fit(feature, y_train_service)
    nb_pred = nb_clf.predict(feature)
    scores = cross_val_score(nb_clf, feature, y_train_service, cv=cv, scoring='accuracy')
    print("Model{} : mean score: {}\n".format(variations[i], np.mean(scores)))
```

```python
def feature_importance(vectorizer, y_train, model):
    #feature importance for lightgbm model
    feature_imp = zip(vectorizer.get_feature_names(), model.feature_importances_)
    sorted_score = sorted(feature_imp, key=lambda x: x[1], reverse=True)
    score = []
    for item in sorted_score:
        score.append([item[0], item[1]])
    return score
```
